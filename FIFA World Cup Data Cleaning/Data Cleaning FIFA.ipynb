```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan':
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# apply fn to contract column and create new columns
new_cols = ['Contract Start', 'Contract End', 'Contract length(Years)']
new_data =  fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
for i in range(len(new_cols)):
    fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i]) #get_loc to get the integer location of the column lable
```

```python
fifa.sample(3)
```

```python
fifa[['Contract','Contract Start','Contract End','Contract length(Years)']].sample(10)
```

```python
# contarct categories
def contract_categories(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# add contract status columns
fifa.insert(fifa.columns.get_loc('Contract length(Years)')+1, 'Contract Status',fifa['Contract'].apply(contract_categories))
fifa.sample(3)
```

```python
# Players height should be in centimeter
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
```

```python
# apply the function to height column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
# rename column
fifa = fifa.rename(columns = {'Height': 'Height(cm)'})
fifa.sample(3)
```

```python
def convert_to_kg(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)
```

```python
fifa['Weight'] = fifa['Weight'].apply(convert_to_kg)
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns = {'Weight': 'Weight(kg)'})
fifa['Weight(kg)'].unique()
```
